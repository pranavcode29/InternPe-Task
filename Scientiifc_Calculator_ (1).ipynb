# Python Program To Develop Scientific Calculator

```python
# Importing Library
import math as m

print("Scientific Calculator")

# Defining Functions to perform various operations

# Define a function to perform addition
def addition(x, y):
    return x + y

# Define a function to perform subtraction
def subtraction(x, y):
    return x - y

# Define a function to perform multiplication
def multiplication(x, y):
    return x * y

# Define a function to perform division
def division(x, y):
    if y == 0:
        return "Error: Cannot divide by Zero"
    return x / y

# Define a function to perform modulus
def modulus(x, y):
    return x % y

# Define a function to perform exponentiation
def power(x, y):
    return x ** y

# Define a function to perform Square
def square(x):
    return x*x

# Define a function to perform square root
def sqrt(x):
    if x < 0:
        return "Error: Cannot calculate square root of a Negative Number"
    return m.sqrt(x)

# Define a function to perform trigonometric functions
def trig(x, func):
    if func == "sin":
        return m.sin(m.radians(x))
    elif func == "cos":
        return m.cos(m.radians(x))
    elif func == "tan":
        return m.tan(m.radians(x))
    else:
        return "Error: Invalid function"

# Define a function to perform logarithmic functions
def log(x, base):
    if x <= 0 or base <= 0 or base == 1:
        return "Error: Invalid input"
    return m.log(x, base)

# Display a menu of options for the user
print("Select operation:")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. modulus")
print("6. Exponentiation")
print("7. Square")
print("8. Square root")
print("9. Trigonometric functions")
print("10. Logarithmic functions")

# Get the user's choice
option = eval(input("What Operation Do You Want To Perform ?: "))

while option <=10:
  if option == 1:
    print("***** Addition Operation *****")
    x = eval(input("Enter First Number: "))
    y = eval(input("Enter Second Number: "))
    result = addition(x,y)
    print("The Addition of {} and {} is {}".format(x, y, result))

  elif option == 2:
    print("***** Subtraction Operation *****")
    x = eval(input("Enter First Number: "))
    y = eval(input("Enter Second Number: "))
    result = subtraction(x,y)
    print("The Subtraction of {} and {} is {}".format(x, y, result))

  elif option == 3:
    print("***** Multiplication Operation *****")
    x = eval(input("Enter First Number: "))
    y = eval(input("Enter Second Number: "))
    result = multiplication(x,y)
    print("The Multiplication of {} and {} is {}".format(x, y, result))

  elif option == 4:
    print("***** Division Operation *****")
    x = eval(input("Enter First Number: "))
    y = eval(input("Enter Second Number: "))
    result = division(x,y)
    print("The Division of {} and {} is {}".format(x, y, result))

  elif option == 5:
    print("***** Modulus Operation *****")
    x = eval(input("Enter First Number: "))
    y = eval(input("Enter Second Number: "))
    result = modulus(x,y)
    print("The Modulus of {} and {} is {}".format(x, y, result))

  elif option == 6:
    print("***** Power Operation *****")
    x = eval(input("Enter First Number: "))
    y = eval(input("Enter Second Number: "))
    result = power(x,y)
    print("The Power of {} and {} is {}".format(x, y, result))

  elif option == 7:
    print("***** Square Operation *****")
    x = eval(input("Enter a Number: "))
    result = square(x)
    print("The Square of {} is {}".format(x, result))

  elif option == 8:
    print("***** Square Root Operation *****")
    x = eval(input("Enter a Number: "))
    result = sqrt(x)
    print("The Square Root of {} is {}".format(x, result))

  elif option == 9:
    print("***** Trigonometric Operations *****")
    num = eval(input("Enter a Number: "))
    func = input("Enter Function (sin, cos, tan): ")
    print(func, "(", num, ")=", trig(num,func))

  elif option == 10:
    print("***** Logarithmic Operations *****")
    num = eval(input("Enter a Number: "))
    base = float(input("Enter base: "))
    print("log_", base, "(", num, ")=", log(num, base))

  else:
    print("Please Choose Correct Operation")

  print()
  new_option = eval(input("Do You Want To Continue Other Operation (Yes-1 or No-0): "))
  if new_option == 1:
    option = eval(input("What Operation Do You Want To Perform ?: "))
  else:
    print("...Thank You So Much...")
    break
```
